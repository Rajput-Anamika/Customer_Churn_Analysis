```python
import numpy as np
```

```python
import pandas as pd
```

```python
df=pd.read_csv('Customer_Churn_Data.csv')
```

```python
df.head().transpose()
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
mapping = {'Male': 1, 'Female': 0}
```

```python
df['gender']=df['gender'].replace(mapping)
```

```python
columns_to_convert = ['Partner','Dependents','PhoneService','PaperlessBilling','Churn']
```

```python
mapping = {'Yes': 1, 'No': 0}
```

```python
df[columns_to_convert] = df[columns_to_convert].replace(mapping)
```

```python
df['TechSupport'].nunique()
```

```python
columns_to_convert = ['TechSupport','StreamingTV','StreamingMovies']
```

```python
mapping = {'Yes': 1, 'No': 0, 'No internet service': 0}
```

```python
df[columns_to_convert] = df[columns_to_convert].replace(mapping)
```

```python
df['Contract'].nunique()
```

```python
mapping = {'One year': 1, 'Month-to-month': 0,'Two year': 2}
```

```python
df['Contract']=df['Contract'].replace(mapping)
```

```python
mapping = {'Electronic check': 1, 'Mailed check': 0,'Bank transfer (automatic)': 2, 'Credit card (automatic)': 3}
```

```python
df['PaymentMethod']=df['PaymentMethod'].replace(mapping)
```

```python
df=pd.get_dummies(df,columns=['OnlineSecurity','OnlineBackup','DeviceProtection','MultipleLines','InternetService'])
```

```python
df['TotalCharges']=df['TotalCharges'].replace(' ',np.nan).astype(float)
```

```python
df.dropna(inplace=True)
```

```python
df.shape
```

```python
df['TotalServices']=df['PhoneService']+df['MultipleLines_Yes']+df['InternetService_DSL']+df['InternetService_Fiber optic']+df['OnlineSecurity_Yes']+df['OnlineBackup_Yes']+df['DeviceProtection_Yes']
```

```python
df['AvgMonthlyCharges']=df['TotalCharges']/df['tenure']
```
